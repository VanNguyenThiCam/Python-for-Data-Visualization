```python
df.head()
```

***Một số toán tử so sánh dùng trong lọc dữ liệu theo điều kiện***

| Toán tử so sánh | Ý nghĩa          |
|:------------:| ---------------:|
| <    | nhỏ hơn           |
| <=          | nhỏ hơn hoặc bằng  |
| > |lớn hơn|
| >= |lớn hơn hoặc bằng|
| == |bằng|
| != |không bằng (khác)|

```python
df['car_type'].value_counts() #xem tổng quan cho biết có bao nhiêu loại và số liệu từng loại
```

```python
car_filter = df['car_type'] == 'Toyota Sienna'
```

```python
car_filter #vị trí nào là Toyota Sienna sẽ trả về true
```

```python
df[car_filter].head() #quy ước là xuất 5 hàng đầu
```

***Cách tiếp cận thứ 2 sử dụng Loc***

```python
df.loc[car_filter,:] #sẽ xuất tất cả Toyota Sienna đã được lọc từ dữ liệu lớn
```

***Toán tử lọc kết hợp nhiều điều kiện***

| Toán tử logic bitwise | Ý nghĩa          |
|:------------:| ---------------:|
| &    | và           |
|    $|$     | hoặc   |
| ^ | Ngược lại với toán tử $|$|
| ~ |không|


```python
df.loc[car_filter & interest_rate_filter]
```
